```python
import pandas as pd
import snscrape.modules.twitter as sntwitter
import itertools
from datetime import datetime,date
```

### Data from 24-06-2017 to 23-06-2018

```python
df_BTC_2017 = pd.DataFrame(itertools.islice(sntwitter.TwitterSearchScraper('#Bitcoin OR #btc OR #bitcoin OR Bitcoin OR btc OR bitcoin lang:en since:2017-06-24 until:2018-06-24 include:nativeretweets min_faves:5 min_retweets:1').get_items(), 1000000))
```

```python
print(df_BTC_2017)
```

```python
df_BTC_2017.to_csv('BTC_2017_NEW')
```

### Data from 24-06-2018 to 23-06-2019
